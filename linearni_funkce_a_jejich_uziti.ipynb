# Linearni funkce a jejich uziti

## üìò √övod
Funkce je p≈ôedpis, kter√Ω ka≈æd√©mu prvku z mno≈æiny D (definiƒçn√≠ obor) p≈ôi≈ôazuje pr√°vƒõ jeden prvek z mno≈æiny H (obor hodnot).  
Zapisujeme: y = f(x)  
Nap≈ô√≠klad: y = 2x + 1

---
### üßÆ 1. Line√°rn√≠ funkce ‚Äì definice, vlastnosti

Definice:  
Line√°rn√≠ funkce je ka≈æd√° funkce tvaru:  
f(x) = ax + b  
kde a, b ‚àà R

- a ‚Äì smƒõrnice p≈ô√≠mky (urƒçuje, jak je p≈ô√≠mka stoupaj√≠c√≠ nebo klesaj√≠c√≠)  
- b ‚Äì posunut√≠ po ose y (pr≈Øseƒç√≠k s osou y)

### üìò Vlastnosti:

‚Äì **Monot√≥nnost** ‚Äì vlastnost funkce, kter√° urƒçuje, zda funkce roste nebo kles√°.  
- Funkce **rostouc√≠**: pokud s rostouc√≠m x roste i hodnota y (graf jde nahoru).  
- Funkce **klesaj√≠c√≠**: pokud s rostouc√≠m x hodnota y kles√° (graf jde dol≈Ø).  
- Funkce **konstantn√≠**: hodnota y se nemƒõn√≠, graf je vodorovn√° p≈ô√≠mka.  

---

‚Ä¢ **Omezenost** ‚Äì ≈ô√≠k√°, zda funkce m√° horn√≠ nebo doln√≠ mez.  
- Funkce je **omezen√° shora**, pokud neexistuje vƒõt≈°√≠ hodnota ne≈æ urƒçit√° mez.  
- Funkce je **omezen√° zdola**, pokud neexistuje men≈°√≠ hodnota ne≈æ urƒçit√° mez.  
- **Line√°rn√≠ funkce** obecnƒõ **nen√≠ omezen√°** ‚Äì m≈Ø≈æe m√≠t libovolnƒõ velk√© i mal√© hodnoty.  

---

‚Ä¢ **Pr≈Øseƒç√≠k s osou x (P‚Çì)** ‚Äì bod, kde graf prot√≠n√° osu x. Plat√≠ y = 0.  
- Urƒç√≠me z rovnice 0 = ax + b ‚Üí x = -b/a  ‚Üí  P‚Çì = (-b/a, 0)  

---

‚Ä¢ **Pr≈Øseƒç√≠k s osou y (P·µß)** ‚Äì bod, kde graf prot√≠n√° osu y. Plat√≠ x = 0.  
- Dosad√≠me x = 0 ‚Üí y = b  ‚Üí  P·µß = (0, b)  

---

Tyto vlastnosti jsou kl√≠ƒçov√© pro pochopen√≠ chov√°n√≠ funkce a pro r√Ωsov√°n√≠ jej√≠ho grafu.

---

### üìâ 2. Graf line√°rn√≠ funkce

Grafem line√°rn√≠ funkce je p≈ô√≠mka.

Pr≈Øseƒç√≠ky s osami:
- osa y: x = 0 ‚Üí y = b ‚Üí Py = (0, b)
- osa x: y = 0 ‚Üí 0 = ax + b ‚Üí x = -b/a ‚Üí Px = (-b/a, 0)


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = 2*x + 1

plt.figure(figsize=(6,4))
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.plot(x, y, label='f(x) = 2x + 1')
plt.scatter(0, 1, color='red', label='Py (0,1)')
plt.scatter(-0.5, 0, color='green', label='Px (-0.5,0)')
plt.title('Graf line√°rn√≠ funkce f(x) = 2x + 1')
plt.legend()
plt.grid(True)
plt.show()
```

---
### ‚úèÔ∏è 3. Urƒçen√≠ p≈ôedpisu z grafu

Pokud zn√°me dva body p≈ô√≠mky A(x‚ÇÅ, y‚ÇÅ) a B(x‚ÇÇ, y‚ÇÇ):

Smƒõrnice: a = (y‚ÇÇ - y‚ÇÅ) / (x‚ÇÇ - x‚ÇÅ)  
Dosazen√≠m jednoho bodu do f(x) = ax + b urƒç√≠me b.


```python
x1, y1 = 1, 3
x2, y2 = 4, 9
a = (y2 - y1) / (x2 - x1)
b = y1 - a*x1
print(f'P≈ôedpis funkce pro body A(1,3) a B(4,9): f(x) = {a}x + {b}')
```

---
### ‚öñÔ∏è 4. Grafick√© ≈ôe≈°en√≠ rovnic a nerovnic

Rovnice f(x) = g(x) ‚Üí pr≈Øseƒç√≠ky graf≈Ø  
Nerovnice f(x) > g(x) ‚Üí oblast, kde je jeden graf nad druh√Ωm


```python
x = np.linspace(-3, 5, 200)
f = 2*x - 1
g = -x + 4

plt.figure(figsize=(6,4))
plt.plot(x, f, label='f(x) = 2x - 1')
plt.plot(x, g, label='g(x) = -x + 4')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.title('Grafick√© ≈ôe≈°en√≠ rovnice f(x) = g(x)')
plt.show()

print('Pr≈Øseƒç√≠k m√° sou≈ôadnice: (5/3, 7/3)')
```

### üåü Shrnut√≠

| **Vlastnost** | **Popis** | **P≈ô√≠klad / Pozn√°mka** |
|----------------|------------|-------------------------|
| **P≈ôedpis** | \( f(x) = ax + b \) | \( a, b \in \mathbb{R} \) |
| **Graf** | P≈ô√≠mka | Urƒçena smƒõrnic√≠ \( a \) a pr≈Øseƒç√≠kem \( b \) |

| **Pr≈Øseƒç√≠k s osou x** | \( P_x = (-b/a, 0) \) | Dosad√≠me \( y = 0 \) |
| **Monot√≥nnost** | Z√°vis√≠ na znam√©nku \( a \) | Viz v√Ω≈°e |
