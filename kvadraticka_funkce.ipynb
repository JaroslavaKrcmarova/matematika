# Kvadratická funkce

## Úvod a motivace
Kvadratické funkce popisují situace jako **vrh míče**, **zisk firmy**, **tvar paraboly antény**.
Míč hozený do vzduchu letí po parabole — tedy grafu kvadratické funkce.

## Definice a vlastnosti
Kvadratická funkce má tvar:
$$ f(x) = ax^2 + bx + c, \quad a \neq 0 $$

**Obor definiční:**
$$ D(f) = (-\infty, +\infty) $$

**Vrchol paraboly:**
$$ x_v = -\frac{b}{2a} $$
$$ y_v = f(x_v) $$

**Obor hodnot:**
- pro $a > 0$:
$$ H(f) = [y_v, +\infty) $$
- pro $a < 0$:
$$ H(f) = (-\infty, y_v] $$

## Grafické znázornění

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)

plt.figure(figsize=(7,4))
plt.plot(x, x**2 - 2*x + 1, label='a > 0')
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.plot(x, -x**2 + 2*x + 1, label='a < 0')
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)
plt.grid(True)
plt.legend()
plt.show()
```

## Řešené příklady

### Příklad 1
Určete vrchol funkce:
$$ f(x) = 2x^2 - 4x + 1 $$

Výpočet:
$$ x_v = -\frac{-4}{2 \cdot 2} = 1 $$
$$ y_v = f(1) = -1 $$

Vrchol:
$$ V = (1, -1) $$

## Shrnutí
- Parabola má tvar $$ f(x) = ax^2 + bx + c $$. 
- Vrchol má souřadnici: $$ x_v = -\frac{b}{2a} $$. 
- Obor hodnot závisí na znaménku $a$. 
- Grafem je parabola otevřená nahoru nebo dolů.
