# üî¢Logaritmy

## üí° Motivace

V matematice se ƒçasto setk√°v√°me s ot√°zkou: **‚ÄûJakou mocninou mus√≠m umocnit ƒç√≠slo, abych dostal jin√© ƒç√≠slo?‚Äú**  
Nap≈ô√≠klad:  
- $2^3 = 8$ ‚Üí ‚Äûƒç√≠m mus√≠m umocnit 2, abych dostal 8?‚Äú Odpovƒõƒè je **3**.

Tuto ot√°zku ≈ôe≈°√≠ **logaritmus**.  
Logaritmy se pou≈æ√≠vaj√≠ v mnoha oblastech:
- p≈ôi pr√°ci s **exponenci√°ln√≠m r≈Østem nebo √∫bytkem** (nap≈ô. radioaktivita, finance, populace),
- p≈ôi mƒõ≈ôen√≠ **hlasitosti** (decibely),
- v **informatice** (slo≈æitost algoritm≈Ø, nap≈ô√≠klad $ O(\log n) $),
- v **chemii a fyzice** (pH, Richterova stupnice zemƒõt≈ôesen√≠).


## üìò Logaritmy ‚Äì definice a vlastnosti

**Definice:**  
Logaritmus ƒç√≠sla $ b $ p≈ôi z√°kladu $ a $ (znaƒç√≠me $ \log_a b $) je ƒç√≠slo \( x \), pro kter√© plat√≠:
$
a^x = b
$
tedy:
$
\log_a b = x
$

### Podm√≠nky:
- $a > 0, a \neq 1$
- $b > 0$

### Z√°kladn√≠ vlastnosti logaritm≈Ø

1. **Souƒçin p≈ôech√°z√≠ na souƒçet:**
   $$\log_a (b \cdot c) = \log_a b + \log_a c$$
    Logaritmus souƒçinu dvou ƒç√≠sel se rovn√° souƒçtu jejich logaritm≈Ø.  (P≈ôev√°d√≠ n√°soben√≠ na sƒç√≠t√°n√≠.)

2. **Pod√≠l p≈ôech√°z√≠ na rozd√≠l:**
   $$
   \log_a \frac{b}{c} = \log_a b - \log_a c
   $$
    Logaritmus pod√≠lu se rovn√° rozd√≠lu logaritm≈Ø.  (P≈ôev√°d√≠ dƒõlen√≠ na odƒç√≠t√°n√≠.)

3. **Mocnina p≈ôech√°z√≠ na souƒçin:**
   $$
   \log_a (b^r) = r \cdot \log_a b
   $$
    Logaritmus mocniny se rovn√° exponentu kr√°t logaritmus z√°kladu.  (Umo≈æ≈àuje ‚Äûvyt√°hnout‚Äú exponent p≈ôed logaritmus.)

4. **Zmƒõna z√°kladu logaritmu:**
   $$
   \log_a b = \frac{\log_c b}{\log_c a}
   $$
    Umo≈æ≈àuje p≈ôepoƒç√≠tat logaritmus s jedn√≠m z√°kladem na logaritmus s jin√Ωm z√°kladem.

## üìà Graf


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 10, 400)
y2 = np.log2(x)
y10 = np.log10(x)
ye = np.log(x)

plt.figure(figsize=(8,5))
plt.plot(x, y2, label=r'$\log_2(x)$')
plt.plot(x, ye, label=r'$\ln(x)$')
plt.plot(x, y10, label=r'$\log_{10}(x)$')

plt.title("Graf logaritmick√Ωch funkc√≠")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

## üß©Vzorce

Logaritmy maj√≠ nƒõkolik z√°kladn√≠ch vztah≈Ø, kter√© se ƒçasto pou≈æ√≠vaj√≠ p≈ôi v√Ωpoƒçtech:

1. **Definiƒçn√≠ vztah:**  
   $$
   a^{\log_a b} = b
   $$
   (Logaritmus je inverzn√≠ k mocninƒõ.)

2. **Jedniƒçka v logaritmu:**  
   $$
   \log_a 1 = 0
   $$
   (Ka≈æd√© ƒç√≠slo umocnƒõn√© na nulu d√°v√° 1.)

3. **Logaritmus z√°kladu:**  
   $$
   \log_a a = 1
   $$
   (Logaritmus sv√©ho z√°kladu je v≈ædy 1.)

4. **Zmƒõna z√°kladu:**  
   $$
   \log_a b = \frac{\log_c b}{\log_c a}
   $$
   (Umo≈æ≈àuje p≈ôepoƒç√≠tat logaritmus s jedn√≠m z√°kladem na jin√Ω.)

5. **Z√°kladn√≠ vlastnosti (p≈ôipomenut√≠):**
   $$
   \begin{align*}
   \log_a (b \cdot c) &= \log_a b + \log_a c \\
   \log_a \frac{b}{c} &= \log_a b - \log_a c \\
   \log_a (b^r) &= r \cdot \log_a b
   \end{align*}
   $$


## üßÆ Uk√°zka ≈ôe≈°en√Ωch p≈ô√≠klad≈Ø

### P≈ô√≠klad 1

**≈òe≈°en√≠:**  
Hled√°me ƒç√≠slo $$x$$, pro kter√© $$2^x = 8$$
Proto $$x = 3$$

### P≈ô√≠klad 2
Vyj√°d≈ôi $x$ z rovnice $5^x = 125$

**≈òe≈°en√≠:**  
Hled√°me ƒç√≠slo $x$, pro kter√© $5^x = 125$.  

V√≠me, ≈æe $125 = 5^3$, tak≈æe dost√°v√°me:
$$
5^x = 5^3
$$

Porovn√°n√≠m exponent≈Ø dostaneme:
$$
x = 3
$$
