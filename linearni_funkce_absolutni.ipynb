
# üìà Line√°rn√≠ funkce s absolutn√≠ hodnotou

## üí° Motivace

Line√°rn√≠ funkce pat≈ô√≠ mezi nejz√°kladnƒõj≈°√≠ typy funkc√≠, kter√© popisuj√≠ **p≈ô√≠mky**.  
V mnoha situac√≠ch ale pot≈ôebujeme, aby funkce mƒõla jen **kladn√© hodnoty** ‚Äì nap≈ô. p≈ôi mƒõ≈ôen√≠ vzd√°lenosti, ceny nebo odchylky.  

V takov√Ωch p≈ô√≠padech se pou≈æ√≠v√° **absolutn√≠ hodnota**, kter√° "obrac√≠" z√°porn√© ƒç√°sti grafu do kladn√©.  
Nap≈ô√≠klad:  
- $f(x) = x$ ‚Üí p≈ô√≠mka  
- $f(x) = |x|$ ‚Üí tvar p√≠smene **V**



## üìò Definice a vlastnosti

**Definice:**  
Line√°rn√≠ funkce s absolutn√≠ hodnotou m√° tvar:
$$
f(x) = |ax + b|
$$

kde:
- $ a, b \in \mathbb{R} $
- v√Ωraz uvnit≈ô absolutn√≠ hodnoty m≈Ø≈æe nab√Ωvat kladn√Ωch i z√°porn√Ωch hodnot  

### üìè Vlastnosti

1. **Definiƒçn√≠ obor:**  
   $$ D(f) = \mathbb{R} $$

2. **Obor hodnot:**  
   $$ H(f) = \langle 0, +\infty) $$
   (Funkce nab√Ωv√° pouze nez√°porn√Ωch hodnot.)

3. **Osa soumƒõrnosti:**  
   Funkce m≈Ø≈æe b√Ωt **osovƒõ soumƒõrn√°** podle p≈ô√≠mky, kde se mƒõn√≠ znam√©nko v√Ωrazu uvnit≈ô absolutn√≠ hodnoty.

4. **Z√°pis po ƒç√°stech:**  
   $$
   f(x) = 
   \begin{cases} 
   ax + b, & \text{pokud } ax + b \ge 0 \\
   -(ax + b), & \text{pokud } ax + b < 0
   \end{cases}
   $$


## üìä Grafick√© zn√°zornƒõn√≠

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
a, b = 1, 0   # parametry funkce
y = np.abs(a*x + b)

plt.figure(figsize=(8,5))
plt.plot(x, y, label=r'$f(x)=|x|$', color='blue')
plt.title("Graf funkce f(x) = |x|")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```


## üßÆ Uk√°zka ≈ôe≈°en√Ωch p≈ô√≠klad≈Ø

### P≈ô√≠klad 1
Urƒçi pr≈Øseƒç√≠ky grafu funkce \( f(x) = |2x - 4| \) s osami sou≈ôadnic.

**≈òe≈°en√≠:**  
Pro pr≈Øseƒç√≠k s osou **x**:
$$
f(x) = 0 \implies |2x - 4| = 0 \implies 2x - 4 = 0 \implies x = 2
$$
‚Üí Pr≈Øseƒç√≠k: **(2, 0)**  

Pro pr≈Øseƒç√≠k s osou **y**:
$$
x = 0 \implies f(0) = |2 \cdot 0 - 4| = 4
$$
‚Üí Pr≈Øseƒç√≠k: **(0, 4)**



### P≈ô√≠klad 2
Zapi≈° funkci \( f(x) = |x - 1| \) bez pou≈æit√≠ absolutn√≠ hodnoty.

**≈òe≈°en√≠:**
$$
f(x) = 
\begin{cases} 
x - 1, & \text{pro } x \ge 1 \\
-(x - 1) = -x + 1, & \text{pro } x < 1
\end{cases}
$$



### P≈ô√≠klad 3
Rozepi≈° funkci \( f(x) = |x - 1| - |-2x + 1| \) bez pou≈æit√≠ absolutn√≠ hodnoty a naƒçrtni jej√≠ pr≈Øbƒõh.

**≈òe≈°en√≠:**

Nejprve urƒç√≠me **nulov√© body, tj. kde se mƒõn√≠ znam√©nko** v√Ωraz≈Ø uvnit≈ô absolutn√≠ch hodnot.

1.   $\quad x - 1 = 0 \quad \Rightarrow  \quad x = 1$
2.   $-2x + 1 = 0 \quad \Rightarrow \quad x = \tfrac{1}{2}$

Tyto body rozdƒõl√≠ osu \( x \) na t≈ôi intervaly:
\[ $(-\infty, \tfrac{1}{2}), \quad (\tfrac{1}{2}, 1), \quad (1, +\infty)$
\]

#### üîπ 1. interval: $ x < \tfrac{1}{2} $

-  $\quad x - 1 < 0 \quad \Rightarrow \quad |x - 1| = -(x - 1) = -x + 1$ 
-  $-2x + 1 > 0 \quad \Rightarrow \quad |-2x + 1| = -2x + 1$ 

Tedy:$f(x) = (-x + 1) - (-2x + 1) = x$

#### üîπ 2. interval: $ \tfrac{1}{2} \le x < 1 $

- $ x - 1 < 0 \Rightarrow |x - 1| = -x + 1 $
- $-2x + 1 < 0 \Rightarrow |-2x + 1| = -(-2x + 1) = 2x - 1 $

Tedy:$f(x) = (-x + 1) - (2x - 1) = -3x + 2$

#### üîπ 3. interval: $x \ge 1 $

- $ x - 1 \ge 0 \Rightarrow |x - 1| = x - 1 $
- $-2x + 1 < 0 \Rightarrow |-2x + 1| = 2x - 1 $

Tedy:$f(x) = (x - 1) - (2x - 1) = -x$

### üìä Z√°pis po ƒç√°stech

$
f(x) = 
\begin{cases}
x, & x < \tfrac{1}{2} \\
-3x + 2, & \tfrac{1}{2} \le x < 1 \\
-x, & x \ge 1
\end{cases}
$


```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 3, 400)
f = np.piecewise(x, 
    [x < 0.5, (x >= 0.5) & (x < 1), x >= 1],
    [lambda x: x, lambda x: -3*x + 2, lambda x: -x]
)

plt.figure(figsize=(8,5))
plt.plot(x, f, label=r'$f(x)=|x-1|-|-2x+1|$', color='purple')
plt.title("Graf funkce f(x) = |x‚àí1| ‚àí |‚àí2x+1|")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=1.5)  # osa x
plt.axvline(0, color='black', linewidth=1.5)  # osa y
plt.show()
```
